Text Vectorization
Convert the prepared text data (questions and answers) into numerical vector representations using Sentence - BERT to capture meaning

```python
from sentence_transformers import SentenceTransformer

#Load a pre trained Sentence - BERT model
model=SentenceTransformer('all-MiniLM-L6-v2')
print("Sentence-BERT model 'all-MiniLM-L6-v2' loaded successfully")
```

```python
import numpy as np

#Generate embeddings for the question column
questions=df_qa['question'].tolist()
question_embeddings=model.encode(questions,show_progress_bar=True)
df_qa['question_embedding']=question_embeddings.tolist()

#Generate embeddings for the 'answer' column
answers=df_qa['answer'].tolist()
answer_embeddings=model.encode(answers,show_progress_bar=True)
df_qa['answer_embedding']=answer_embeddings.tolist()

print("EMbeddings generated for 'question' and 'asnwer' colunms")
print(df_qa.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_best_match(user_query, question_embeddings, df_qa, model):
  #Vectorize the user query
  user_query_embedding=model.encode([user_query])

  #Calculate cosine similarity between user query and question embeddings
  similarities=cosine_similarity(user_query_embedding,question_embeddings)

  #Find the index of the highest similarity
  best_match_index=np.argmax(similarities)

  #Retrive and return the coressponding anwer
  return df_qa.loc[best_match_index]['answer']

print("find_best_match funcition defined successfully.")
```

```python
sample_query="How to prevent from a power cut?"

best_answer=find_best_match(sample_query,question_embeddings,df_qa,model)

print(f"User Query: {sample_query}")
print(f"Retrieved Answer: {best_answer}")
```

```python
def chat_with_bot():
  print("Welcome to the ELectricity Authority Chatbot\n")
  print("Ask me anything about power outages, billings, services or safety. Type 'quit' to exit.\n")

  while True:
    user_query=input("You: ")
    if user_query.lower() =="quit":
      print("Chatbot: Thankyou for chatting with us . Goodbye!")
      break

    #Retrive the best asnwer using the defined funtion and existing data
    retrieved_asnwer= find_best_match(user_query, question_embeddings, df_qa, model)
    print(f"CHatbot: {retrieved_asnwer}\n")

#Invoke the chatbot function to start the interaction
# chat_with_bot()
```
